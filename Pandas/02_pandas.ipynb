# Handling Missing Values

### 1. Detect Missing Values

| Function |  Explanation |
|----------|-----------------------|
| `df.isna()` | Returns True for missing (NA) values. |
| `df.isnull()` | Same as isna(); checks for missing values. |
| `df.notna()` | Returns True for non-missing values. |
| `df.isna().sum()` | Counts missing values in each column. |


```python
import pandas as pd
df = pd.read_csv('Titanic-Dataset.csv')
df.head()
```

```python
df.isna()
```

```python
df.isnull()
```

```python
df.notna()
```

```python
df.isna().sum()
```

### 2. Drop Missing Values

| Function | Explanation |
|----------|-----------------------|
| `df.dropna()` | Removes all rows with any missing values. |
| `df.dropna(axis=1)` | Removes columns that contain missing values. |
| `df.dropna(thresh=N)` | Keeps rows with at least N non-missing values. |

```python
df.dropna()
```

```python
df.dropna(axis = 1)
```

```python
df.dropna(thresh = 2)
```

### 3. Fill Missing Values

| Function | Explanation |
|----------|-----------------------|
| `df.fillna(value)` | Fills missing values with a specified value. |
| `df['col'].fillna(df['col'].mean())` | Fills missing values using the column mean. |
| `df['col'].fillna(df['col'].median())` | Fills missing using the column median. |
| `df['col'].fillna(df['col'].mode()[0])` | Fills missing using the most common value (mode). |
| `df.ffill()` | Forward fill: uses previous row's value. |
| `df.bfill()` | Backward fill: uses next row's value. |

```python
df.fillna(-1)
```

```python
df['Age'].fillna(df['Age'].mean())
```

```python
df['Age'].fillna(df['Age'].median())
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df.ffill()
```

```python
df.bfill()
```

### 4.  Replace Missing or Special Values

| Function | Explanation |
|----------|-----------------------|
| `df.replace("?", pd.NA)` | Replaces custom placeholders (like '?') with NA. |
| `df.replace([None, "NA"], pd.NA)` | Converts custom missing markers to NA. |

```python
df.replace('?', pd.NA)
```

```python
df.replace([None, 'NA'], pd.NA)
```
